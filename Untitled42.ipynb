```python
import numpy as np
import pandas as pd

# =========================================================
# 0) 排序（非常关键：确保shift(-1)就是下一根bar）
# =========================================================
sort_cols = [c for c in ['SecuCode', 'TradingDay', 'TimeStart', 'TimeEnd', 'TickTime'] if c in df.columns]
if not sort_cols:
    raise ValueError("df里需要能排序的列，比如 SecuCode/TradingDay/TimeStart(TimeEnd) 或 TickTime.")

df = df.sort_values(sort_cols).reset_index(drop=True)

# 分组：通常bar数据用 (SecuCode, TradingDay)，tick也一样
g = df.groupby(['SecuCode', 'TradingDay'], sort=False)

# =========================================================
# 1) mid / spread / microprice / book_state（去重：mid只算一次）
# =========================================================
bid1 = df['BidPrice1_last']
ask1 = df['AskPrice1_last']
bv1  = df['BidVolume1_last'] if 'BidVolume1_last' in df.columns else np.nan
av1  = df['AskVolume1_last'] if 'AskVolume1_last' in df.columns else np.nan

# mid: 双边 -> 均值；单边 -> 非0那一侧；都0 -> NaN
df['mid'] = np.where(
    (bid1 > 0) & (ask1 > 0),
    0.5 * (bid1 + ask1),
    np.where(bid1 > 0, bid1, np.where(ask1 > 0, ask1, np.nan))
)

# spread：只在双边盘口时定义
df['spread'] = np.where(
    (bid1 > 0) & (ask1 > 0),
    (ask1 - bid1),
    np.nan
)

# microprice：双边且量有效才算；否则退化为 mid
den = (bv1 + av1)
df['microprice'] = np.where(
    (bid1 > 0) & (ask1 > 0) & (den > 0),
    (ask1 * bv1 + bid1 * av1) / den,
    df['mid']
)

df['microprice_minus_mid'] = df['microprice'] - df['mid']

# 盘口状态：0=双边正常，1=仅bid，2=仅ask，NaN=都没
df['book_state'] = np.select(
    [
        (bid1 > 0) & (ask1 > 0),
        (bid1 > 0) & (ask1 <= 0),
        (bid1 <= 0) & (ask1 > 0),
    ],
    [0, 1, 2],
    default=np.nan
)

# =========================================================
# 2) label: mid 的下一根bar log return（t -> t+1）
#    这里就是你要的 “mid return 对应 t+1 return”
# =========================================================
# 防止log(0)/log(NaN)：mid<=0 直接给 NaN
mid_pos = df['mid'].where(df['mid'] > 0)

df['log_mid_ret_t1'] = g['mid'].transform(
    lambda x: np.log(x.shift(-1)) - np.log(x)
)


```

```python
import numpy as np
import pandas as pd
import inspect
import importlib
import Factor as F
importlib.reload(F)

# 1) 排序
df = df.sort_values(["SecuCode","TradingDay","TimeEnd","TimeStart"]).reset_index(drop=True)

# 2) y
y_col = "log_mid_ret_t1"
if y_col not in df.columns:
    g = df.groupby(["SecuCode","TradingDay"], sort=False)
    df[y_col] = np.log(df["mid"].astype(float)) - np.log(g["mid"].shift(1).astype(float))


key_cols = ["SecuCode","TradingDay","TimeStart","TimeEnd"]
for c in key_cols:
    df[c] = df[c].astype("int64")  # 强制保住

num_cols = [c for c in df.select_dtypes(include=["number"]).columns if c not in key_cols]
df[num_cols] = df[num_cols].astype("float32")

def compute_factors_001_070_fast(df, i=20, prefix="f_", j_default=4, k_default=1.0):
    out = {}
    for k in range(1, 71):
        fname = f"factor_{k:03d}"
        if not hasattr(F, fname):
            continue
        func = getattr(F, fname)
        sig = inspect.signature(func)

        kwargs = {}
        if "i" in sig.parameters: kwargs["i"] = i
        if "j" in sig.parameters: kwargs["j"] = j_default
        if "k" in sig.parameters: kwargs["k"] = k_default

        out[f"{prefix}{k:03d}"] = func(df, **kwargs).astype("float32")

    factor_df = pd.DataFrame(out, index=df.index)
    return factor_df, list(out.keys())

factor_df, factor_cols = compute_factors_001_070_fast(df, i=20, prefix="f_", j_default=4, k_default=1.0)
print("computed factors:", len(factor_cols))

index_cols = ["SecuCode","TradingDay","TimeStart","TimeEnd"]
df_train = pd.concat([df[index_cols + [y_col]], factor_df], axis=1)

df_train = df_train.dropna(subset=[y_col])

print(df_train["TradingDay"].max())
df_train

```

```python
# y 列
y_col = "log_mid_ret_t1"

# 1) 找所有 factor
factor_cols = [c for c in df_train.columns if c.startswith("f_")]

# 2) 找缺失率最高的前 4 个 factor
drop_factors = (
    df_train[factor_cols]
    .isna()
    .mean()
    .sort_values(ascending=False)
    .head(4)
    .index
    .tolist()
)

print("Dropped factors:", drop_factors)

# 3) drop 这 4 个 factor
df_train = df_train.drop(columns=drop_factors)

# 4) 重新拿剩余 factor
factor_cols_clean = [c for c in df_train.columns if c.startswith("f_")]

# 5) 对 y + 剩余 factor 删除缺失值
df_train = df_train.dropna(subset=factor_cols_clean + [y_col])

print("Remaining factors:", len(factor_cols_clean))
print("Final rows:", len(df_train))

```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb

# ===== 0) TradingDay 先转成 int，避免奇怪的 day =====
df_train = df_train.copy()
df_train["TradingDay"] = df_train["TradingDay"].astype(int)

# ===== 1) 划分：最后两年(约 2*252 个交易日)做 test =====
all_days = np.sort(df_train["TradingDay"].unique())
assert len(all_days) > 2 * 252, "交易日数量不够做最后两年 test"

test_start_day = all_days[-2 * 252]
print("Test start day:", test_start_day)

df_tr = df_train[df_train["TradingDay"] < test_start_day].copy()
df_te = df_train[df_train["TradingDay"] >= test_start_day].copy()
print("train rows:", len(df_tr), "test rows:", len(df_te))

# ===== 2) 特征 & 标签 =====
y_col = "log_mid_ret_t1"
feature_cols = [c for c in df_train.columns if c.startswith("f_")]

X_train = df_tr[feature_cols]
y_train = df_tr[y_col].astype(float)

X_test  = df_te[feature_cols]
y_test  = df_te[y_col].astype(float)

# ===== 3) LightGBM Dataset =====
train_set = lgb.Dataset(X_train, label=y_train, free_raw_data=False)
test_set  = lgb.Dataset(X_test,  label=y_test, reference=train_set, free_raw_data=False)

# ===== 4) 参数 =====
params = {
    "objective": "regression",
    "metric": "rmse",
    "learning_rate": 0.05,
    "num_leaves": 64,
    "max_depth": -1,
    "min_data_in_leaf": 2000,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "lambda_l1": 1.0,
    "lambda_l2": 5.0,
    "verbosity": -1,
    "seed": 42
}

# ===== 5) ✅ 用 callbacks 替代 early_stopping_rounds / verbose_eval =====
callbacks = [
    lgb.early_stopping(stopping_rounds=50),
    lgb.log_evaluation(period=50),
]

model = lgb.train(
    params=params,
    train_set=train_set,
    num_boost_round=1000,
    valid_sets=[train_set, test_set],
    valid_names=["train", "test"],
    callbacks=callbacks
)

# ===== 6) 预测 & RMSE =====
y_pred = model.predict(X_test, num_iteration=model.best_iteration)

rmse = float(np.sqrt(np.mean((y_test.values - y_pred) ** 2)))
print("Test RMSE:", rmse)

# ===== 7) IC（日内截面相关：每天 pred vs y）=====
df_te_eval = df_te[["TradingDay", y_col]].copy()
df_te_eval["pred"] = y_pred

def _daily_ic(g):
    # Spearman 更稳：对高频 noisy label 友好
    return g["pred"].corr(g[y_col], method="spearman")

ic = df_te_eval.groupby("TradingDay", sort=False).apply(_daily_ic)
print("IC mean:", float(ic.mean()))
print("IC std :", float(ic.std()))
print("IC IR  :", float(ic.mean() / (ic.std() + 1e-12)))
```
